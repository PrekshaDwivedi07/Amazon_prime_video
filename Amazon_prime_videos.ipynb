### Dataset Description
This dataset contains information about 124,347 entries related to movies and TV shows, including metadata about the titles, their cast and crew, ratings, genres, and production details. It is likely designed for analyzing streaming content or entertainment media, such as from Amazon Prime or a similar platform.

###Column Descriptions

person_id: A unique numeric identifier assigned to each person (actor, director, etc.).

id: A unique string identifier for each title (could be internal or platform-specific).

name: The name of the person (actor, director, etc.) involved in the content.

character: The name of the character portrayed by the person. This is available mostly for actors and is missing in around 13% of the records.

role: Specifies the role of the person in the production — such as actor, director, writer, etc.

title: The name of the movie or TV show.

type: Indicates whether the content is a movie or a TV show.

description: A short synopsis or summary of the content. Almost all entries have this, with only a few missing.

release_year: The year in which the movie or show was released.

age_certification: The age rating (like PG, R, or TV-MA) that guides appropriate viewing age. However, over 50% of entries have missing age certification.

runtime: The duration of the movie or episode in minutes.

genres: The categories or genres assigned to the content, such as Action, Drama, Comedy, etc.

production_countries: Lists the countries involved in the production of the content.

seasons: Applicable mainly to TV shows, this column records the number of seasons. Since most entries are movies, this field is missing for over 90% of the records.

imdb_id: The IMDb ID used to look up the content on IMDb.

imdb_score: The IMDb rating on a scale from 0 to 10. Some records are missing this score, possibly due to the title not being listed or not reviewed.

imdb_votes: Number of user votes received on IMDb for that title.

tmdb_popularity: Popularity score derived from The Movie Database (TMDb), which can help gauge how trending a title is.

tmdb_score: Rating score from TMDb, also out of 10. Around 10% of the entries are missing this data.

###Box plot of imdb_score

Box plots are plotted to visualize the distributionn,spread of data and identify the outliers in data
